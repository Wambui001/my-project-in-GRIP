# Task 4 EXPLORATORY DATA ANALYSIS USING PYTHON

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import warnings
warnings.filterwarnings('ignore')

df=pd.read_csv('globalterrorismdb_0718dist.csv',encoding='ISO-8859-1')
```

```python
print(df.columns)
df['iyear'].dtype
```

```python
plt.figure(figsize=(12,6))
sns.countplot(x=df['iyear'],data=df)
plt.xticks(rotation=90)
plt.title('Terrorist Attacks Over the Years')
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.countplot(x=df['country_txt'],data= df,order=df['country_txt'].value_counts().index[:10])
plt.xticks(rotation=90)
plt.title('Top 10 Countries with Most Terrorist Attacks ')
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.countplot(x='attacktype1_txt',data=df,order=df['attacktype1_txt'].value_counts().index)
plt.xticks(rotation=90)
plt.title('Types of Terrorist Attacks ')
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.countplot(x='weaptype1_txt',data=df,order=df['weaptype1_txt'].value_counts().index)
plt.xticks(rotation=90)
plt.title('Types of Weapons Used in Attacks')
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.countplot(x=df['iyear'],data=df)
plt.xticks(rotation=90)
plt.title('Terrorist Attacks Over the Years')
plt.savefig('terrorist_attacks_over_years.png')
```
